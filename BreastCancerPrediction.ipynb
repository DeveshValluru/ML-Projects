```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and PreProcessing

```python
breast_cancer_data = pd.read_csv('/content/breast_cancer_data.csv')
print(breast_cancer_data)
```

```python
breast_cancer_data.head()
```

```python
breast_cancer_data.shape
```

```python
breast_cancer_data.info()
```

```python
breast_cancer_data.describe()
```

```python
breast_cancer_data.isnull().sum()
```

```python
breast_cancer_data.replace({'diagnosis':{'M':0,'B':1}}, inplace=True)

breast_cancer_data.head()
```

```python
breast_cancer_data.groupby('diagnosis').mean()
```

```python
X = breast_cancer_data.drop(columns=['diagnosis','id','Unnamed: 32'],axis=1)
Y = breast_cancer_data['diagnosis'].astype(int)
```

```python
print(X)
```

```python
print(Y)
```

splitting train test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

Model Training 

LogisticRegression

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Model Evaluation

```python
train_data_prediction = model.predict(X_train)
train_score = accuracy_score(Y_train,train_data_prediction)
print('accuracy score of train data is:', train_score)
```

```python
test_data_prediction = model.predict(X_test)
test_score = accuracy_score(Y_test,test_data_prediction)
print('accuracy score of test data is:', test_score)
```

Saving the model

```python
import pickle
```

```python
filename = 'breast_cancer_model.sav'
pickle.dump(model,open(filename,'wb'))
```
